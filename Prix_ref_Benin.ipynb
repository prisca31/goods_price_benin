Problème : dans la colonne "Numéro" de mon fichier /content/prix_benin.csv j'ai des entrées avec du texte et j'ai des entrées avec des nombres mais qui ne sont pas lus comme tels

Je trouve une façon de ne garder que les entrées où il y a les chiffres

```python
import pandas as pd
import numpy as np

# Charger le fichier CSV
df = pd.read_csv('/content/prix_benin.csv')

# Définir une fonction pour essayer de convertir en nombre
def est_un_nombre(x):
    if isinstance(x, str):  # vérifier si la valeur est une chaîne de caractères
        try:
            float(x.replace(' ', ''))  # supprimer les espaces et essayer de convertir en nombre
            return True
        except ValueError:
            return False
    else:  # si la valeur n'est pas une chaîne de caractères, elle est déjà un nombre
        return True

# Appliquer la fonction à la colonne "Numéro"
df['est_un_nombre'] = df['Numéro'].apply(est_un_nombre)

# Filtrer les lignes où la colonne "Numéro" contient des nombres
df_filtre = df[df['est_un_nombre']]

# Supprimer la colonne 'est_un_nombre' qui n'est plus nécessaire
df_filtre = df_filtre.drop('est_un_nombre', axis=1)
```

```python
# Enregistrer le résultat dans un nouveau fichier CSV
df_filtre.to_csv('resultat.csv', index=False)
```

J'attribue les catégories à chaque produit selon son id_key

```python
# Définir les catégories et les plages d'id_key correspondantes
categories = [
    (1, 115, 'mobiliers de bureau'),
    (116, 317, 'materiels et equipements de bureau'),
    (318, 905, 'fournitures de bureau et materiels pedagogiques'),
    (906, 1053, 'materiels informatiques et accessoires'),
    (1054, 1227, 'consommables informatiques'),
    (1228, 3301, 'automobiles et pieces detachees'),
    (3302, 3375, 'matériels roulant à deux et à trois roues'),
    (3376, 3889, 'produits alimentaires et equipements de restauration'),
    (3890, 4638, 'produits de sante humaine'),
    (4639, 4874, 'produits d’entretien'),
    (4875, 6505, 'genie-civil, genie electrique et infrastructures routieres'),
    (6506, 6537, 'groupes electrogenes et accessoires'),
    (6538, 6735, 'equipements sportifs'),
    (6736, 7508, 'equipements et autres produits agricoles'),
    (7509, 7540, 'location de vehicules'),
    (7541, 7549, 'prestations intellectuelles de services'),
    (7550, 7561, 'prestations de services d’entretien et de securisation des batiments publics')
]

# Créer une fonction pour attribuer une catégorie à chaque ligne
def attribuer_categorie(id_key):
    for debut, fin, categorie in categories:
        if debut <= id_key <= fin:
            return categorie
    return 'inconnu'  # si l'id_key ne correspond à aucune catégorie

# Appliquer la fonction à chaque ligne du DataFrame
df_clean['categorie'] = df_clean['id_key'].apply(attribuer_categorie)
```
